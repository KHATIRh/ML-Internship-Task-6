# Q.5: Identify the most common interests among users, which is vital for segmenting the audience and tailoring ad campaigns to match user preferences.

```python
from collections import Counter
interests_series = df['Top Interests'].str.split(',').explode()

interest_counts = Counter(interests_series)

top_10_interests = interest_counts.most_common(10)

top_10_interests_df = pd.DataFrame(top_10_interests, columns=['Interest', 'Count'])

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Interest', data=top_10_interests_df, palette='coolwarm')
plt.title('Top 10 Most Common Interests Among Users')
plt.xlabel('Count')
plt.ylabel('Interest')

plt.show()
```

# Q.6: You can now segment users into distinct groups for targeted ad campaigns. Segmentation can be based on various criteria, such as:
# ->Demographics: Age, Gender, Income Level, Education Level
# ->Behavioural: Time Spent Online, Likes and Reactions, CTR, Conversion Rates
# ->Interests: Aligning ad content with the top interests identified
# To implement user profiling and segmentation, apply clustering techniques or develop personas based on the combination of these attributes.


```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

df=pd.read_csv('user_profiles_for_ads.csv')

df.fillna(0, inplace=True)

def age_range_to_midpoint(age_range):
    if '-' in age_range:
        start, end = age_range.split('-')
        return (int(start) + int(end)) / 2
    elif '+' in age_range:
        start = age_range.split('+')[0]
        return int(start) + 5
    return int(age_range)

df['Age'] = df['Age'].apply(age_range_to_midpoint)

features = ['Age', 'Gender', 'Income Level', 'Education Level', 
            'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)', 
            'Likes and Reactions', 'Click-Through Rates (CTR)', 
            'Conversion Rates', 'Ad Interaction Time (sec)']

df['Gender'] = df['Gender'].astype('category').cat.codes
df['Income Level'] = df['Income Level'].astype('category').cat.codes
df['Education Level'] = df['Education Level'].astype('category').cat.codes

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=features)

print("Cluster Centers:")
print(cluster_centers_df)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Time Spent Online (hrs/weekday)', hue='Cluster', data=df, palette='viridis')
plt.title('User Segmentation based on Age and Time Spent Online (Weekdays)')
plt.xlabel('Age')
plt.ylabel('Time Spent Online (hrs/weekday)')
plt.legend(title='Cluster')
plt.show()
```

# Q.7: Now compute the mean values of the numerical features and the mode for categorical features within each cluster to get a sense of their defining characteristics.


```python
df['Age'].fillna(0, inplace=True)
df.fillna(0, inplace=True)

def age_range_to_midpoint(age_range):
    if isinstance(age_range, str):
        if '-' in age_range:
            start, end=age_range.split('-')
            return (int(start) + int(end)) / 2
        elif '+' in age_range:
            start=age_range.split('+')[0]
            return int(start)+5
        return int(age_range)
    return age_range

df['Age']=df['Age'].apply(age_range_to_midpoint)

features=['Age', 'Gender', 'Income Level', 'Education Level', 'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates', 'Ad Interaction Time (sec)']

df['Gender']=df['Gender'].astype('category').cat.codes
df['Income Level']=df['Income Level'].astype('category').cat.codes
df['Education Level']=df['Education Level'].astype('category').cat.codes

scaler=StandardScaler()
df_scaled=scaler.fit_transform(df[features])

kmeans=KMeans(n_clusters=5, random_state=42)
df['Cluster']=kmeans.fit_predict(df_scaled)

numerical_features=['Age', 'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates', 'Ad Interaction Time (sec)']
cluster_means=df.groupby('Cluster')[numerical_features].mean()

categorical_features=['Gender', 'Income Level', 'Education Level']
cluster_modes=df.groupby('Cluster')[categorical_features].agg(lambda x: x.mode().iloc[0])

cluster_profiles=pd.concat([cluster_means, cluster_modes], axis=1)

print('Cluster Profiles:')
print(cluster_profiles)
```
